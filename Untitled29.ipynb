```python
#Network Graph
import networkx as nx
import matplotlib.pyplot as plt

edges = [('A', 'B'), ('B', 'C'), ('C', 'A'), ('D', 'A'), ('D', 'C')]
G = nx.DiGraph(edges)
nx.draw(G, with_labels=True, node_size=700, node_color='skyblue', font_color='black', arrowsize=20, alpha=0.7)
plt.show()
```

```python
#Dependency Graph
import networkx as nx
import matplotlib.pyplot as plt

# Sample dataset with dependencies
dependencies = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A'), ('E', 'A')]

# Create a directed graph
G = nx.DiGraph()

# Add dependencies to the graph
G.add_edges_from(dependencies)

# Draw the network graph
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', arrowsize=20)

# Show the graph
plt.title('Dependency Graph')
plt.show()
```

```python
#boxplot
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)
ds1 = np.random.normal(100,20,220)
ds2 = np.random.normal(80,20,200)
ds3 = np.random.normal(60,35,220)
ds4 = np.random.normal(50,40,200)
ds = [ds1,ds2,ds3,ds4]
figure = plt.figure(figsize=(10,7))
ax = figure.add_axes([0,0,1,1])
bp = ax.boxplot(ds)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
colors = x + y
plt.scatter(x,y,c=colors,cmap='viridis',edgecolor='black')
cbar = plt.colorbar()
cbar.set_label('sum of x and y')
plt.title('same colloured values')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Plot pairplot
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.show()
```
